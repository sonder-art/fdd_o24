```python
import numpy as np
import time
from joblib import Parallel, delayed
import math
import multiprocessing as mp

# Crear un array grande
arr = np.random.rand(10**7)

# For loop computacionalmente pesado
start = time.time()
result = []
for x in arr:
    result.append(x**2)
result = np.array(result)
end = time.time()

print(f"Tiempo con for loop: {end - start} segundos")
```

```python
# Usando numpy para vectorizar la operaci칩n
start = time.time()
result_np = arr**2
end = time.time()

print(f"Tiempo con numpy: {end - start} segundos")
```

```python
# Definir funci칩n que har치 la operaci칩n
def compute_square(x):
    return x**2

# Usar joblib para paralelizar el for loop
start = time.time()
result_joblib = Parallel(n_jobs=-1)(delayed(compute_square)(x) for x in arr)
end = time.time()

print(f"Tiempo con joblib: {end - start} segundos")
```

```python
# Definir funci칩n para calcular el cuadrado
def compute_square_mp(x):
    return x**2

# Usar multiprocessing para paralelizar
if __name__ == '__main__':
    start = time.time()
    pool = mp.Pool(mp.cpu_count())
    result_mp = pool.map(compute_square_mp, arr)
    pool.close()
    end = time.time()

    print(f"Tiempo con multiprocessing: {end - start} segundos")
```
