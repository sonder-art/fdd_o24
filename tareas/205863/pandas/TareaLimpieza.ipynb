```python
#Casteo de columnas, eliminación de columnas no útiles y manejo de los NaNs
loans['term'] = loans['term'].str.replace(' months', '').astype(int)
loans.rename(columns={'term_(months)': 'term(months)'}, inplace=True)
loans['grade'] = loans['grade'].astype(str)
#Todas son letras, es más fácil de manejar un string
loans['emp_title'] = loans['emp_title'].astype(str)
loans['emp_title'].str.upper()
#A string porque todas son cadenas y lo ponemos en mayusculas para facilitar queries  tener orden
loans['verification_status'].astype(str)
loans['verification_status'].str.upper()
#Todas son cadenas
loans = loans.drop('issue_d', axis=1)
loans = loans.drop('pymnt_plan', axis=1)
loans = loans.drop('hardship_flag', axis=1)
#Todas son iguales no tiene caso tener esa columna para el análisis
loans = loans.drop('desc', axis=1)
loans = loans.dropna(axis=1, how='all')
# Eliminar todas las columnas que están completamente vacías
loans['emp_length'] = loans['emp_length'].str.replace('years', '')
#Ya sabemos que son años, si fueran meses serían irrelevantes.
#  Aun así vendría en la descripción de la limpieza
loans['title'] = loans['title'].astype(str)
loans['title'].str.upper()
loans['zip_code'] = loans['zip_code'].astype(str)
loans['addr_state'] = loans['addr_state'].astype(str)
#Todas son cadenas 
loans = loans.fillna(0)
#Rellenamos los NaN con 0 para que todas las columnas esten llenas, 
# es más fácil manejar el 0 que el NaN. Escogí este método para enfocar el análisis 
# más en ver cuanta gente debe en vez de la cantidad. Yo asumí que si un dato es NaN 
# está así es porque el usuario no debe nada y por eso no se registró
loans = loans.drop(loans.columns[94:], axis=1)
loans = loans.drop(['pub_rec_bankruptcies', 'tax_liens'], axis=1)
#Eliminamos las columnas de la 94 en adelante porque la gran mayoría están vacias 
#y son excepciones las que no o el valor en toda la columna es la misma



```
